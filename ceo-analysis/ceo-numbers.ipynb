```python
import pandas as pd
```

```python
data = pd.read_json('https://interactives.dallasnews.com/data-store/2018/2018-06-ceo-pay-tracker.json')
```

```python
data
```

Finding average of ceo pay in 2017 - need to exclude null values

```python
ceo_complete = data.dropna(subset=['ceopay2017'])
ceo_complete['ceopay2017'].mean()
```

```python
ceo_complete
```

Finding average of employee pay in 2017 - excluding null values

```python
employee_complete = data.dropna(subset=['medianpay'])
employee_complete['medianpay'].mean()
```

Finding median ceo pay in 2017 with excluded null values 

```python
ceo_complete['ceopay2017'].median()
```

Finding average ceo pay by sector

```python
cd = ceo_complete[ceo_complete.sectorfilter == 'consumer-discretionary']
communications = ceo_complete[ceo_complete.sectorfilter == 'communications']
financials = ceo_complete[ceo_complete.sectorfilter == 'financials']
utilities = ceo_complete[ceo_complete.sectorfilter == 'utilities']
energy = ceo_complete[ceo_complete.sectorfilter == 'energy']
# energy = ceo_complete[ceo_complete.sectorfilter == 'energy']
materials = ceo_complete[ceo_complete.sectorfilter == 'materials']
hc = ceo_complete[ceo_complete.sectorfilter == 'health-care']
industrials = ceo_complete[ceo_complete.sectorfilter == 'industrials']
technology = ceo_complete[ceo_complete.sectorfilter == 'technology']
cs = ceo_complete[ceo_complete.sectorfilter == 'consumer-staples']
```

```python
sectors = [cd, communications, financials, utilities, energy, materials, hc, industrials, technology, cs]
```

```python
for sector in sectors:
  print(sector.ceopay2017.mean())
```

Find average of ceo to median pay ratios - exclude null values first 

```python
ratio_complete = data.dropna(subset=['employeeratio'])
ratio_complete.employeeratio.mean()
```

Find median of ceo to median pay ratios

```python
ratio_complete.employeeratio.median()
```

Find average pay ratio by sector - exclude null ratio values first 

```python
r_cd = ratio_complete[ratio_complete.sectorfilter == 'consumer-discretionary']
r_communications = ratio_complete[ratio_complete.sectorfilter == 'communications']
r_financials = ratio_complete[ratio_complete.sectorfilter == 'financials']
r_utilities = ratio_complete[ratio_complete.sectorfilter == 'utilities']
r_energy = ratio_complete[ratio_complete.sectorfilter == 'energy']
r_materials = ratio_complete[ratio_complete.sectorfilter == 'materials']
r_hc = ratio_complete[ratio_complete.sectorfilter == 'health-care']
r_industrials = ratio_complete[ratio_complete.sectorfilter == 'industrials']
r_technology = ratio_complete[ratio_complete.sectorfilter == 'technology']
r_cs = ratio_complete[ratio_complete.sectorfilter == 'consumer-staples']
```

```python
r_sectors = [r_cd, r_communications, r_financials, r_utilities, r_energy, r_materials, r_hc, r_industrials, r_technology, r_cs]
```

```python
for r_sector in r_sectors:
  print(r_sector.employeeratio.mean())
```

Find median pay ration by sector

```python
for r_sector in r_sectors:
  print(r_sector.employeeratio.median())
```
